# Lemmatization

### Wordnet Lemmatizer

```python
import nltk
nltk.download('wordnet')
```

```python
from nltk.stem import WordNetLemmatizer
```

```python
lemmatizer = WordNetLemmatizer()
```

### Pos(Part of speech) tag for words 
1. n → Noun
2. v → Verb
3. a → Adjective
4. r → Adverb

POS Tagging:
The process of assigning a grammatical label (noun, verb, adjective, etc.) to each word in a sentence.
Based on pos tag we got the root word of the specific word 

```python
lemmatizer.lemmatize("going",pos='v')
```

```python
words = ["eating","eats","eaten","writing","writes","programming","programs","history","finally","finalize"]
```

```python
for word in words:
    print(word+"---->"+lemmatizer.lemmatize(word,pos='v'))
```
